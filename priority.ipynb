```python
def handle_interaction(**kwargs):
    if  kwargs.get('type') == 'click' and not aoi_widget.value:
              
        
        latlon = kwargs.get('coordinates')
     
        
        Map.default_style = {"cursor": "wait"}
        xy = ee.Geometry.Point(latlon[::-1])
        roi = Protected_areas.filterBounds(xy)
        
        print(latlon)
        
        
        #Load datasets
        elevation=ee.Image("USGS/SRTMGL1_003").select("elevation").clip(roi)
        
        slope=ee.Terrain.slope(elevation).clip(roi)
        
        bio1=ee.Image("WORLDCLIM/V1/BIO").select('bio01').clip(roi)
        bio12=ee.Image("WORLDCLIM/V1/BIO").select('bio12').clip(roi)
        PercentTree_cover=ee.ImageCollection("MODIS/006/MOD44B").select('Percent_Tree_Cover').median().clip(roi)
        Percent_NonVegetated=ee.ImageCollection("MODIS/006/MOD44B").select('Percent_NonVegetated').median().clip(roi)
        PercentNon_tree=ee.ImageCollection("MODIS/006/MOD44B").select('Percent_NonTree_Vegetation').median().clip(roi)
        landcover=ee.ImageCollection("ESA/WorldCover/v100").first().clip(roi)
        download=timeseries.value
        
        #Reclassify
        if slope_classify.value=="Ascending":
            
            
            slopere_reclass = ee.Image(1) \
              .where(slope.gt(0.00).And(slope.lte(0.04)), 2) \
              .where(slope.gt(0.004).And(slope.lte(0.08)), 3) \
              .where(slope.gt(0.1).And(slope.lte(0.4)), 4) \
              .where(slope.gt(0.4).And(slope.lte(0.6)), 5).rename('slope')
        else:
             slopere_reclass = ee.Image(1) \
              .where(slope.lte(0.6).And(slope.gt(0.4)), 2) \
              .where(slope.lte(0.4).And(slope.gt(0.1)), 3) \
              .where(slope.lte(0.08).And(slope.gt(0.04)), 4) \
              .where(slope.lte(0.04).And(slope.gt(0.00)), 5).rename('slope')
            
        if elevation_classify.value=="Ascending":
            
            elevation_reclass=ee.Image(1) \
              .where(elevation.gt(100).And(elevation.lte(200)), 2) \
              .where(elevation.gt(200).And(elevation.lte(300)), 3) \
              .where(elevation.gt(300).And(elevation.lte(400)), 4) \
              .where(elevation.gt(400).And(elevation.lte(500)), 5)\
              .where(elevation.gt(500).And(elevation.lte(600)), 6)\
              .where(elevation.gt(600).And(elevation.lte(700)), 7).rename('elevation')
        else:
            elevation_reclass=ee.Image(1) \
              .where(elevation.lte(700).And(elevation.gt(600)), 2) \
              .where(elevation.lte(600).And(elevation.gt(500)), 3) \
              .where(elevation.lte(500).And(elevation.gt(400)), 4) \
              .where(elevation.lte(400).And(elevation.gt(300)), 5)\
              .where(elevation.lte(300).And(elevation.gt(200)), 6)\
              .where(elevation.lte(200).And(elevation.gt(100)), 7).rename('elevation')
            
        
        
        if annual_temperature_classify.value=="Ascending":
            
                  
            bio1_reclass= ee.Image(1) \
                .where(bio1.gt(-10).And(bio1.lte(30)), 2) \
                .where(bio1.gt(30).And(bio1.lte(70)), 3) \
                .where(bio1.gt(70).And(bio1.lte(110)), 4) \
                .where(bio1.gt(110).And(bio1.lte(150)), 5)\
                .where(bio1.gt(150).And(bio1.lte(190)), 6)\
                .where(bio1.gt(190).And(bio1.lte(230)), 7)\
                .where(bio1.gt(230).And(bio1.lte(270)), 8)\
                .where(bio1.gt(270).And(bio1.lte(320)), 9)\
                .where(bio1.gte(320), 10).rename('annual_temperature')
        else:
            bio1_reclass= ee.Image(1) \
                .where(bio1.lte(320).And(bio1.gt(270)), 2) \
                .where(bio1.lte(270).And(bio1.gt(230)), 3) \
                .where(bio1.lte(230).And(bio1.gt(190)), 4) \
                .where(bio1.lte(190).And(bio1.gt(150)), 5)\
                .where(bio1.lte(150).And(bio1.gt(110)), 6)\
                .where(bio1.lte(110).And(bio1.gt(70)), 7)\
                .where(bio1.lte(70).And(bio1.gt(30)), 8)\
                .where(bio1.lte(30).And(bio1.gt(-10)), 9)\
                .where(bio1.lte(-10), 10).rename('annual_temperature')
            
        
            
        if annual_precipitation_classify.value=="Ascending":
            
            
            bio12_reclass=ee.Image(1) \
              .where(bio12.gt(1266).And(bio12.lte(2533)), 2) \
              .where(bio12.gt(2533).And(bio12.lte(3801)), 3) \
              .where(bio12.gt(3801).And(bio12.lte(5067)), 4) \
              .where(bio12.gt(5067).And(bio12.lte(7600)), 5)\
              .where(bio12.gt(7600).And(bio12.lte(8867)), 6)\
             .where(bio12.gt(8867).And(bio12.lte(10134)), 7)\
              .where(bio12.gt(10134).And(bio12.lte(11401)), 8).rename('annual_precipitation')
        else:
             bio12_reclass=ee.Image(1) \
              .where(bio12.lte(11401).And(bio12.gt(10134)), 2) \
              .where(bio12.lte(10134).And(bio12.gt(8867)), 3) \
              .where(bio12.lte(8867).And(bio12.gt(7600)), 4) \
              .where(bio12.lte(7600).And(bio12.gt(5067)), 5)\
              .where(bio12.lte(5067).And(bio12.gt(3801)), 6)\
              .where(bio12.lte(3801).And(bio12.gt(2533)), 7)\
              .where(bio12.lte(2533).And(bio12.gt(1266)), 8).rename('annual_precipitation')
                
        if tree_cover_classify.value=="Ascending": 
            
            
        
           PercentTree_cover_reclass=ee.Image(1) \
              .where(PercentTree_cover.gt(10).And(PercentTree_cover.lte(20)), 2) \
              .where(PercentTree_cover.gt(20).And(PercentTree_cover.lte(30)), 3) \
              .where(PercentTree_cover.gt(30).And(PercentTree_cover.lte(40)), 4) \
              .where(PercentTree_cover.gt(40).And(PercentTree_cover.lte(50)), 5)\
              .where(PercentTree_cover.gt(50).And(PercentTree_cover.lte(60)), 6)\
              .where(PercentTree_cover.gt(60).And(PercentTree_cover.lte(70)), 7)\
              .where(PercentTree_cover.gt(70).And(PercentTree_cover.lte(80)), 8)\
              .where(PercentTree_cover.gt(80).And(PercentTree_cover.lte(90)), 9)\
              .where(PercentTree_cover.gt(90).And(PercentTree_cover.lte(100)), 10).rename('percent_tree_cover')
        else:
            
            PercentTree_cover_reclass=ee.Image(1) \
              .where(PercentTree_cover.lte(100).And(PercentTree_cover.gt(90)), 2) \
              .where(PercentTree_cover.lte(90).And(PercentTree_cover.gt(80)), 3) \
              .where(PercentTree_cover.lte(80).And(PercentTree_cover.gt(70)), 4) \
              .where(PercentTree_cover.lte(70).And(PercentTree_cover.gt(60)), 5)\
              .where(PercentTree_cover.lte(60).And(PercentTree_cover.gt(50)), 6)\
              .where(PercentTree_cover.lte(60).And(PercentTree_cover.gt(50)), 7)\
              .where(PercentTree_cover.lte(40).And(PercentTree_cover.gt(30)), 8)\
              .where(PercentTree_cover.lte(40).And(PercentTree_cover.gt(30)), 9)\
              .where(PercentTree_cover.lte(20).And(PercentTree_cover.gt(10)), 10).rename('percent_tree_cover')
            
        if nonvegetatedwi_classify.value=="Ascending":
            
        
            Percent_NonVegetated_reclass=ee.Image(1) \
              .where(Percent_NonVegetated.gt(10).And(Percent_NonVegetated.lte(20)), 2) \
              .where(Percent_NonVegetated.gt(20).And(Percent_NonVegetated.lte(30)), 3) \
              .where(Percent_NonVegetated.gt(30).And(Percent_NonVegetated.lte(40)), 4) \
              .where(Percent_NonVegetated.gt(40).And(Percent_NonVegetated.lte(50)), 5)\
              .where(Percent_NonVegetated.gt(50).And(Percent_NonVegetated.lte(60)), 6)\
              .where(Percent_NonVegetated.gt(60).And(Percent_NonVegetated.lte(70)), 7)\
              .where(Percent_NonVegetated.gt(70).And(Percent_NonVegetated.lte(80)), 8)\
              .where(Percent_NonVegetated.gt(80).And(Percent_NonVegetated.lte(90)), 9)\
              .where(Percent_NonVegetated.gt(90).And(Percent_NonVegetated.lte(100)), 10).rename('percent_non_vegetated')
        else:
            
              
            Percent_NonVegetated_reclass=ee.Image(1) \
              .where(Percent_NonVegetated.lte(100).And(Percent_NonVegetated.gt(90)), 2) \
              .where(Percent_NonVegetated.lte(90).And(Percent_NonVegetated.gt(80)), 3) \
              .where(Percent_NonVegetated.lte(80).And(Percent_NonVegetated.gt(70)), 4) \
              .where(Percent_NonVegetated.lte(70).And(Percent_NonVegetated.gt(60)), 5)\
              .where(Percent_NonVegetated.lte(60).And(Percent_NonVegetated.gt(50)), 6)\
              .where(Percent_NonVegetated.lte(50).And(Percent_NonVegetated.gt(40)), 7)\
              .where(Percent_NonVegetated.lte(40).And(Percent_NonVegetated.gt(30)), 8)\
              .where(Percent_NonVegetated.lte(30).And(Percent_NonVegetated.gt(20)), 9)\
              .where(Percent_NonVegetated.lte(20).And(Percent_NonVegetated.gt(10)), 10).rename('percent_non_vegetated')
            
        
        if non_tree_classify.value=="Ascending":
            
            PercentNon_tree_reclass=ee.Image(1) \
              .where(PercentNon_tree.gt(10).And(PercentNon_tree.lte(20)), 2) \
              .where(PercentNon_tree.gt(20).And(PercentNon_tree.lte(30)), 3) \
              .where(PercentNon_tree.gt(30).And(PercentNon_tree.lte(40)), 4) \
              .where(PercentNon_tree.gt(40).And(PercentNon_tree.lte(50)), 5)\
              .where(PercentNon_tree.gt(50).And(PercentNon_tree.lte(60)), 6)\
              .where(PercentNon_tree.gt(60).And(PercentNon_tree.lte(70)), 7)\
              .where(PercentNon_tree.gt(70).And(PercentNon_tree.lte(80)), 8)\
              .where(PercentNon_tree.gt(80).And(PercentNon_tree.lte(90)), 9)\
              .where(PercentNon_tree.gt(90).And(PercentNon_tree.lte(100)), 10).rename('percentnon_tree')
        else:
            PercentNon_tree_reclass=ee.Image(1) \
              .where(PercentNon_tree.lte(100).And(PercentNon_tree.gt(90)), 2) \
              .where(PercentNon_tree.lte(90).And(PercentNon_tree.gt(80)), 3) \
              .where(PercentNon_tree.lte(80).And(PercentNon_tree.gt(70)), 4) \
              .where(PercentNon_tree.lte(70).And(PercentNon_tree.gt(60)), 5)\
              .where(PercentNon_tree.lte(60).And(PercentNon_tree.gt(50)), 6)\
              .where(PercentNon_tree.lte(50).And(PercentNon_tree.gt(40)), 7)\
              .where(PercentNon_tree.lte(40).And(PercentNon_tree.gt(30)), 8)\
              .where(PercentNon_tree.lte(30).And(PercentNon_tree.gt(20)), 9)\
              .where(PercentNon_tree.lte(20).And(PercentNon_tree.gt(10)), 10).rename('percentnon_tree')
            
        
        landcover_reclass= ee.Image(1) \
          .where(landcover.eq(10), 2) \
          .where(landcover.eq(20), 3) \
          .where(landcover.eq(30), 4) \
          .where(landcover.eq(40), 5) \
          .where(landcover.eq(50), 6) \
          .where(landcover.eq(60), 7) \
          .where(landcover.eq(70), 8) \
          .where(landcover.eq(80), 9) \
          .where(landcover.eq(90), 10) \
          .where(landcover.eq(95), 11).rename('landcover')
        
        elev_text=elewi.value +'*'+'elevation +'
        slop_text=slopewi.value+'*'+'slope +'
        temp_text=bio1wi.value+'*'+'annual_temperature +'
        preci_text=bio12wi.value+"*" + 'annual_precipitation +'
        tree_cover_text=tree_coverwi.value +"*"+'percent_tree_cover +'
        non_vegetated_text=nonvegetatedwi.value + "*"+'percent_non_vegetated +'
        non_tree_text=non_treewi.value + "*"+'percentnon_tree +'
        landcover_text=landcoverwi.value + "*" + 'landcover '
        model=elev_text+slop_text+temp_text+preci_text+tree_cover_text+non_vegetated_text+non_tree_text+landcover_text
        
        model1=slopere_reclass.clip(roi)
        model2=elevation_reclass.clip(roi)
        model3=bio1_reclass.clip(roi)
        model4=bio12_reclass.clip(roi)
        model5=PercentTree_cover_reclass.clip(roi)
        model6=Percent_NonVegetated_reclass.clip(roi)
        model7=PercentNon_tree_reclass.clip(roi)
        model8=landcover_reclass.clip(roi)
        stacked_image=model1.addBands(model2).addBands(model3).addBands(model4).addBands(model5).addBands(model6).addBands(model7).addBands(model8)
        
        
        
        result=stacked_image.expression(model, {
        'slope': stacked_image.select('slope'),
        'elevation': stacked_image.select('elevation'),
        'annual_temperature': stacked_image.select('annual_temperature'),
        'annual_precipitation': stacked_image.select('annual_precipitation'),
        'percent_tree_cover': stacked_image.select('percent_tree_cover'),
        'percent_non_vegetated': stacked_image.select('percent_non_vegetated'),
        'percentnon_tree': stacked_image.select('percentnon_tree'), 
        'landcover': stacked_image.select('landcover')
            })
    
        
        def mean(image):
                                                              
            image = ee.Image(image)
            meandict=image.reduceRegion(**{
                                'reducer': ee.Reducer.mean(),
                                'geometry': roi,
                                'scale': 30,
                                'maxPixels': 1e9
                                 })

            new_property= image.set('mean', meandict)
            return new_property
        
        
        #get mean
        slope_mean=mean(slope)
        elevation_mean=mean(elevation)
        bio1_mean=mean(bio1)
        bio12_mean=mean(bio12)
        PercentTree_cover_mean=mean(PercentTree_cover)
        Percent_NonVegetated_mean=mean(Percent_NonVegetated)
        PercentNon_tree_mean=mean(PercentNon_tree)
        
        
        
        
        
        #build dataframe
      
        
        elevation_coll=elevation_mean.get('mean').getInfo()
        elevation_final=pd.DataFrame.from_dict([elevation_coll])
        
       
        slope_coll=slope_mean.get('mean').getInfo()
        slope_final=pd.DataFrame.from_dict([slope_coll])
        
        bio1_coll=bio1_mean.get('mean').getInfo()
        bio1_final=pd.DataFrame.from_dict([bio1_coll])
        
        bio12_coll=bio12_mean.get('mean').getInfo()
        bio12_final=pd.DataFrame.from_dict([bio12_coll])
        
        PercentTree_cover_coll=PercentTree_cover_mean.get('mean').getInfo()
        PercentTree_cover_final=pd.DataFrame.from_dict([PercentTree_cover_coll])
        
        Percent_NonVegetated_coll= Percent_NonVegetated_mean.get('mean').getInfo()
        Percent_NonVegetated_final=pd.DataFrame.from_dict([Percent_NonVegetated_coll])
        
        PercentNon_tree_coll=PercentNon_tree_mean.get('mean').getInfo()
        PercentNon_tree_final=pd.DataFrame.from_dict([PercentNon_tree_coll])
        
        #merge dataframes
        
        dataframe_list=[slope_final, elevation_final,bio1_final, bio12_final, PercentTree_cover_final,Percent_NonVegetated_final, PercentNon_tree_final]
        merged=pd.concat(dataframe_list, axis=1)
        
       
       
        
        
            
        
        collections= ee.ImageCollection("MODIS/006/MOD44B").select('Percent_Tree_Cover').filterBounds(roi).filterDate('2000-01-01', '2021-02-3')
        
        clipped=collections.map(lambda img: img.clip(roi))
        dates_original = clipped.aggregate_array("system:time_start").getInfo()
        day = [datetime.datetime.fromtimestamp(i/1000).strftime('%Y-%m-%d') for i in (dates_original)]
        mean_treecover_series=clipped.map(mean)
        final_mean_coll=mean_treecover_series.aggregate_array('mean').getInfo()
        mean_dataframe=pd.DataFrame(final_mean_coll, index=day)
        mean_dataframe.index = pd.to_datetime(mean_dataframe.index, format="%Y/%m/%d")
        mean_dataframe.sort_index(ascending = True, inplace = True)
            
     
        
                
               
      

        with output_widget:
            
            output_widget.clear_output()
            
           
                            
            Map.addLayer(result, {'palette':['006400', 'ffbb22','ffff4c', 'f096ff','fa0000','f0f0f0' ]}, 'weigthed overlay')
            Map
        
            
            
        with out1:
                   
            out1.clear_output()
            print(merged)
            print('Selected location:', latlon)
            fig = px.scatter(mean_dataframe['Percent_Tree_Cover'],  width=900, height=500)
            fig.show()
          
        if download: 
            
            
        
            
            out_dir1 = os.path.join(os.path.expanduser('~'), 'Downloads')
            out_name1 =   geemap.random_string() + '.csv'
            out_csv1 = os.path.join(out_dir1, out_name1)
            if not os.path.exists(out_dir1):
                
                 
                os.makedirs(out_dir1)
           
            mean_dataframe.to_csv(out_csv1)
            link1 = geemap.create_download_link(
            out_csv1, title="Click here to download the timeseries data: ")
            display(link1)
       
       
       
                 
            
Map.default_style = {'cursor': 'pointer'}            
Map.on_interaction(handle_interaction)
```
